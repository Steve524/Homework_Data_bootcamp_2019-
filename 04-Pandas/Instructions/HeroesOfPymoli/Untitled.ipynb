```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
players_data = len(purchase_data['SN'].value_counts())
```

```python
Summary_Table = pd.DataFrame({"Total Players": [players_data]})

Summary_Table
```

```python
unique_items = len(purchase_data["Item Name"].unique())
avg_purchase = purchase_data["Price"].astype(float).mean()
total_purchase = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].astype(float).sum() #.map("${:,.2f}".format)
```

```python
pd.options.display.float_format = '${:,.2f}'.format
```

```python
Summary_Table = pd.DataFrame({"Number of Unique Items": [unique_items],
                             "Average Purchase Price": [avg_purchase],
                             "Total Number of Purchases": [total_purchase],
                             "Total Revenue": [total_revenue]})

Summary_Table
```

```python
#gender_stats = (purchase_data['Gender'].value_counts())
#gender_stats = pd.DataFrame({
    #"Total Count": purchase_data['Gender'].value_counts(),
    #"Percentage of Players": purchase_data['Gender'].sum()/players_data
#})
gender_stats = purchase_data.groupby('Gender')
unique_gender = gender_stats.nunique()['SN']

gender_percen = unique_gender / players_data * 100

gender_yomp = pd.DataFrame({
    "Total Count": unique_gender,
    "Percentage of Players": gender_percen
})
gender_yomp.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"%{:.2f}"})    
#pd.DataFrame(gender_stats)
```

```python
purchase_count = gender_stats['Purchase ID'].count()
avg_purchase = gender_stats["Price"].mean()
total_purchase = gender_stats["Price"].sum()
avg_total_purchase = total_purchase / unique_gender

gender_stats_purchase = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase,
    "Total Purchase Value": total_purchase,
    "Avg Total Purchase per Person": avg_total_purchase
})

gender_stats_purchase
    
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


age_grouped = purchase_data.groupby("Age Group")
unique_age = age_grouped.nunique()['SN']

age_percen = unique_age / players_data * 100

age_yomp = pd.DataFrame({
    "Total Count": unique_age,
    "Percentage of Players": age_percen
})

age_yomp.style.format({"Percentage of Players":"{:.2f}%"})
```

```python
purchase_count = age_grouped['Purchase ID'].count()
avg_purchase = age_grouped["Price"].mean()
total_purchase = age_grouped["Price"].sum()
avg_total_purchase = total_purchase / unique_age

age_stats_purchase = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase,
    "Total Purchase Value": total_purchase,
    "Avg Total Purchase per Person": avg_total_purchase
})

age_stats_purchase
    
```

```python
SN_top_players = purchase_data.groupby('SN')
purchase_count = SN_top_players['Purchase ID'].count()
avg_purchase = SN_top_players["Price"].mean()
total_purchase = SN_top_players["Price"].sum()

top_players_yomp = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase,
    "Total Purchase Value": total_purchase,
})

top_players_yomp.sort_values(["Total Purchase Value"], ascending=False).head()
```

```python
top_items = purchase_data.groupby('Item Name')
purchase_count = top_items['Purchase ID'].count()
avg_purchase = top_items["Price"].count()
total_purchase = top_items["Price"].sum()

top_players_yomp = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase,
    "Total Purchase Value": total_purchase,
})
top_players_yomp.sort_values(["Total Purchase Value"], ascending=False).head()
```
